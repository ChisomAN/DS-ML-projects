## Convert Reviews to Sentences

```r
h1_sentences <- get_sentences(hotel1$Review)
h2_sentences <- get_sentences(hotel2$Review)
h3_sentences <- get_sentences(hotel3$Review)
```

## Sentiment Scoring

```r
h1_sentiment <- get_sentiment(h1_sentences, method = "syuzhet")
h2_sentiment <- get_sentiment(h2_sentences, method = "syuzhet")
h3_sentiment <- get_sentiment(h3_sentences, method = "syuzhet")
```

## Simple Trendlines

```r
simple_plot(h1_sentiment)
simple_plot(h2_sentiment)
simple_plot(h3_sentiment)
```

## Moving Averages and Rescaling

```r
h1_ma <- rollmean(h1_sentiment, 5, fill = NA)
h2_ma <- rollmean(h2_sentiment, 5, fill = NA)
h3_ma <- rollmean(h3_sentiment, 5, fill = NA)

h1_rescaled <- rescale_x_2(h1_sentiment)
h2_rescaled <- rescale_x_2(h2_sentiment)
h3_rescaled <- rescale_x_2(h3_sentiment)

plot(h1_rescaled$x, h1_ma, type="l", col="red", main="Hotel 1 - Rescaled MA")
plot(h2_rescaled$x, h2_ma, type="l", col="blue", main="Hotel 2 - Rescaled MA")
plot(h3_rescaled$x, h3_ma, type="l", col="green", main="Hotel 3 - Rescaled MA")
```

## DCT Smoothing

```r
h1_dct <- get_dct_transform(h1_sentiment, low_pass_size = 5, x_reverse_len = 100)
h2_dct <- get_dct_transform(h2_sentiment, low_pass_size = 5, x_reverse_len = 100)
h3_dct <- get_dct_transform(h3_sentiment, low_pass_size = 5, x_reverse_len = 100)

plot(h1_dct, type="l", col="red", main="DCT Hotel 1")
plot(h2_dct, type="l", col="blue", main="DCT Hotel 2")
plot(h3_dct, type="l", col="green", main="DCT Hotel 3")

length(h1_dct)
length(h2_dct)
length(h3_dct)
```

## Combined DCT Comparison

```r
plot(h1_dct, type="l", col="red", ylim=c(min(h1_dct, h2_dct, h3_dct), max(h1_dct, h2_dct, h3_dct)), main="DCT Comparison")
lines(h2_dct, col="blue")
lines(h3_dct, col="green")
legend("topright", legend=c("Hotel 1", "Hotel 2", "Hotel 3"), col=c("red", "blue", "green"), lty=1)
```

## Cosine Similarity (via Correlation)

```r
cor(h1_dct, h2_dct)
cor(h1_dct, h3_dct)
cor(h2_dct, h3_dct)
```

## Interpretation of Results

From the sentiment summary statistics:
- **Hotel 2** has the highest **mean sentiment score (0.6461)** and **median (0.5)**, indicating overall more favorable reviews.
- **Hotel 1** follows with a **mean of 0.4689**, showing moderately positive feedback.
- **Hotel 3** has the lowest **mean (0.2888)** and lowest maximum score, suggesting weaker sentiment intensity.

The neutral sentiment percentages were:
- Hotel 1: ~34%
- Hotel 2: ~26%
- Hotel 3: ~27%

This shows Hotel 1’s reviews contain more neutral statements, possibly pointing to more balanced or mixed experiences.

The cosine similarity (correlation) values:
- Hotel 1 and Hotel 2: -0.70 (strong negative correlation)
- Hotel 1 and Hotel 3: -0.88 (very strong negative correlation)
- Hotel 2 and Hotel 3: 0.45 (moderate positive correlation)

This indicates that Hotels 1 and 3 diverge significantly in sentiment trajectory, while Hotels 2 and 3 share a somewhat similar emotional pattern.

**Managerial Insight:** Hotel 2, though positive, should be monitored due to its higher variance and polarity shifts. Hotel 3 requires improvement in customer service or facility, given its lower sentiment values.

## Ethical Reflection

When conducting sentiment analysis, ethical safeguards are critical. One risk is **data omission bias**, where unfavorable reviews are excluded—intentionally or accidentally—leading to skewed positivity. If Hotel 2 were to remove the bottom 10% of reviews, its already high average would falsely climb, distorting renovation or investment decisions.

Another concern is **algorithmic transparency**. Without knowing how syuzhet assigns polarity, decision-makers may misinterpret model outcomes. Responsible use includes documenting all preprocessing steps, validating the model on known benchmarks, and using multiple sentiment scoring tools where feasible (Jockers, 2017).

**Bias mitigation**, **user privacy**, and **responsible interpretation** are essential to ethical sentiment mining (Zhou et al., 2020). Ensuring transparency helps maintain integrity and stakeholder trust.

## Conclusion

Despite Hotel 2 having the highest mean sentiment, Hotel 1’s steadier trend and fewer extremes make it the most stable in terms of guest satisfaction. Hotel 3’s low scores indicate opportunities for strategic improvement.

By combining quantitative sentiment scores with visual trend analysis and ethical consideration, this report offers actionable insights for hotel managers. Renovation priority could be allocated toward Hotel 3, while Hotel 2 might benefit from customer service training. Hotel 1 can maintain current operations with minimal intervention.

Ultimately, sentiment mining is a powerful decision-support tool when applied transparently, accurately, and ethically.
